# Array Properties and Manipulation

1. Array Shape and Dimensions:

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6], [4, 5, 6]])


print(arr.shape)  
```

2. Reshaping Arrays

```python
# Reshape the array to have 6 rows and 1 column
reshaped_arr = arr.reshape(6, 1)
print(reshaped_arr)  
```

3. Array Concatenation (Vertically):

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Concatenate vertically (stack arrays on top of each other)
combined_arr = np.vstack((arr1, arr2))
print(combined_arr)  # Output: [[1 2 3] [4 5 6]]
```

4. Array Concatenation (Horizontally):

```python
# Concatenate horizontally (place arrays side-by-side)
combined_arr = np.hstack((arr1, arr2))
print(combined_arr)  
```

5. Array Splitting:

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
split_arrs = np.split(arr, 2, axis=0)  # Split along rows (axis=0)
print(split_arrs[0])  
print(split_arrs[1])
```

6. Linear Algebra Operations:

```python
# Matrix multiplication
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
product = np.dot(matrix1, matrix2) 
print(product)  


```

```python
# Matrix transpose
transposed = np.transpose(matrix1)
print(transposed)  # Output: [[1 3], [2 4]]

```
